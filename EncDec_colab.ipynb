```python
import torch
import torch.nn as nn

class ChessEncoder(nn.Module):
  def __init__(self):
    super().__init__()

    # CNN Encoder for board representation
    self.cnn = nn.Sequential(
      nn.Conv2d(1, 32, kernel_size=3, padding=1),
      nn.ReLU(),
      nn.MaxPool2d(2),
      nn.Conv2d(32, 64, kernel_size=3, padding=1),
      nn.ReLU(),
      nn.MaxPool2d(2),
      nn.Flatten()
    )

    # Embedding layers for move ID, move number, and turn
    self.move_embedding = nn.Embedding(len(UCI_MOVES), 32)
    self.turn_embedding = nn.Embedding(2, 4)  # Assuming two players

    # RNN for processing concatenated representation
    self.rnn = nn.LSTM(293, 64,num_layers = 4)  # Input size = 256 (CNN) + 32  + 4 + 1

  def forward(self, board, move_id, move_number, turn):
    board_embedding = self.cnn(board.unsqueeze(1))
    concat_embedding = torch.cat([
      board_embedding,
      self.move_embedding(move_id).unsqueeze(0),
      torch.tensor(move_number).view(-1, 1),
      self.turn_embedding(turn).unsqueeze(0)
    ], dim=1)
    rnn_output, _ = self.rnn(concat_embedding)

    return rnn_output

class Mapping(nn.Module):

    def __init__(self, embed_size, device):
        super(Mapping, self).__init__()
        self.embed_size = embed_size
        self.transformer_encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=embed_size, nhead=16, dim_feedforward=embed_size * 4, dropout=0.08,
                                      device=device), num_layers=5).to(device)
        self.mapper = nn.Linear(embed_size, 768).to(device)
        self.init_weights()

    def forward(self, img_embedded, train_mode=False):
        x = self.transformer_encoder(img_embedded)
        x = self.mapper(x)
        x = x.view(*([-1, 1, 768] if train_mode else [1, 768]))
        return x

    def init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_in', nonlinearity='relu')
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.LayerNorm):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)


class ChessDecoder(nn.Module):
  def __init__(self,tokenizer,device):
    super().__init__()
    self.decoder = GPT2LMHeadModel.from_pretrained("gpt2")
    self.tokenizer = tokenizer
    self.tokenizer.pad_token = self.tokenizer.eos_token



  def forward(self, encoder_outputs, attention_mask=None):
    # Use relevant arguments from GPT2LMHeadModel for teacher forcing and generation
    return self.decoder(inputs_embeds=encoder_outputs,attention_mask=attention_mask).logits


class ChessCommentaryGenerator(nn.Module):
  def __init__(self, tokenizer, device):
      super().__init__()
      self.device = device
      self.encoder = ChessEncoder().to(device)
      self.mapper = Mapping(64, device).to(device)
      self.decoder = ChessDecoder(tokenizer,device).to(device)
      self.optimizer = torch.optim.Adam(self.parameters(), lr=0.001)
      self.criterion = nn.CrossEntropyLoss(ignore_index=self.decoder.tokenizer.pad_token_id)


  def train_forward(self, board, move_id, move_number, turn, comment,attention_mask):
    self.optimizer.zero_grad()
    x , x_mask = comment[:,:-1].to(self.device),attention_mask[:,:-1].to(self.device)
    y = comment[:,1:].to(self.device)
    board = board.to(self.device)
    move_id = move_id.to(self.device)
    move_number = move_number.to(self.device)
    turn = turn.to(self.device)
    encoder_output = self.encoder(board, move_id, move_number, turn).to(self.device)
    mapper_output = self.mapper(encoder_output,train_mode = True).to(self.device)
    text_emb = self.decoder.decoder.transformer.wte(x).to(self.device)
    # print(text_emb.shape)
    # text_emb = text_emb.view(1,text_emb.shape[-2],text_emb.shape[-1])
    x = torch.concat([mapper_output,text_emb],dim=1).to(self.device)
    x_mask = torch.concat([torch.ones(x_mask.shape[0],1).to(self.device),x_mask],dim=1).to(self.device)
    pos_emb = self.decoder.decoder.transformer.wpe(torch.arange(x.shape[1]).to(self.device)).to(self.device)
    pos_emb = pos_emb.expand_as(x).to(self.device)
    x = x+ pos_emb
    res = self.decoder(x,attention_mask=x_mask).to(self.device)
    res = torch.softmax(res,dim=2).to(self.device)
    loss = self.criterion(res[:,1:,:].reshape(-1,res.shape[-1]),y.reshape(-1)).to(self.device)
    loss.backward()
    self.optimizer.step()
    return loss
```
